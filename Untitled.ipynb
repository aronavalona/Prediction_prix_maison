```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
data = fetch_california_housing(as_frame=True)
df = data.frame
df.head()
```

```python
df.info()
```

```python
df.describe()
df.isnull().sum()
```

```python
sns.histplot(df['MedHouseVal'], bins=50, kde=True)
plt.title("Distribution des prix des maisons")
plt.xlabel("Prix (en 100k$)")
plt.show()
```

```python
df.corr()['MedHouseVal'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrélation")
plt.show()
```

```python
X = df.drop('MedHouseVal', axis=1)
y = df['MedHouseVal']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.002, random_state=1000)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"MAE : {mae:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"R² Score : {r2:.2f}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Prix réel")
plt.ylabel("Prix prédit")
plt.title("Prix réels vs Prix prédits")
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red')  # ligne diagonale
plt.show()
```

```python
errors = y_test - y_pred
sns.histplot(errors, bins=50, kde=True)
plt.title("Distribution des erreurs")
plt.xlabel("Erreur")
plt.show()
```

```python
coeff_df = pd.DataFrame(model.coef_, index=X.columns, columns=["Coefficient"])
coeff_df.sort_values(by="Coefficient", ascending=False)
```

```python
import joblib
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Supposons que df est déjà nettoyé
X = df.drop('MedHouseVal', axis=1)
y = df['MedHouseVal']  # ou np.log(df['MedHouseVal']) si tu veux log-transformer

# Standardisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Entraînement
model = LinearRegression()
model.fit(X_train, y_train)

# Sauvegarde
joblib.dump(model, 'model.pkl')
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(X.columns.tolist(), 'features.pkl')
```
